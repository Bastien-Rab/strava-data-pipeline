```python
import pandas as pd
from sqlalchemy import create_engine

# Connexion PostgreSQL
engine = create_engine("postgresql://airflow_user:TON_MDP@localhost:5432/strava")

# Charger les stats mensuelles
df_stats = pd.read_sql("SELECT * FROM fct_activities_stats ORDER BY mois DESC", engine)

df_stats.head(60)
```

```python
import matplotlib.pyplot as plt

df_pivot = df_stats.pivot(index="mois", columns="type", values="total_km")
df_pivot.plot(kind="bar", stacked=True, figsize=(14, 6))
plt.title("📏 Distance totale par mois et type d'activité")
plt.ylabel("Kilomètres")
plt.xlabel("Mois")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
